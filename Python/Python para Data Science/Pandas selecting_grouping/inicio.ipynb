# Agrupamento de dados

```python
emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns
```

```python
emissoes_gases_colunas_info = list(emissoes_gases.loc[:,'Nível 1 - Setor':'Produto'].columns)
emissoes_gases_colunas_info
```

```python
emissoes_gases.loc[:,1970:2021].columns
```

```python
emissoes_gases_colunas_values = list(emissoes_gases.loc[:,1970:2021].columns)
emissoes_gases_colunas_values
```

```python
emissoes_gases.melt(id_vars=emissoes_gases_colunas_info, value_vars=emissoes_gases_colunas_values,var_name='Ano', value_name='Emissão')
```

```python
emissoes_por_ano = emissoes_gases.melt(id_vars=emissoes_gases_colunas_info, value_vars=emissoes_gases_colunas_values,var_name='Ano', value_name='Emissão')
```

```python
emissoes_por_ano.groupby('Gás')
```

```python
emissoes_por_ano.groupby('Gás').groups
```

```python
emissoes_por_ano.groupby('Gás').get_group('CO2 (t)')
```

```python
emissoes_por_ano.groupby('Gás')[['Emissão']].sum()
```

```python
emissao_por_gas = emissoes_por_ano.groupby('Gás')[['Emissão']].sum().sort_values('Emissão', ascending= False)
```

```python
emissao_por_gas.plot(kind='barh', figsize=(10,6))
```

```python
emissoes_de_co2 = emissao_por_gas.iloc[0:9]
emissoes_de_co2
```

```python
print(f'Emissão total de 1970 a 2021 no Brasil: {float(emissao_por_gas.iloc[0:9].sum()/emissao_por_gas.sum().iloc[0])*100:.2f}%')
```
