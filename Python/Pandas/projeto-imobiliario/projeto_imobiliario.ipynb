# Tratando e filtrando os dados

## Lidando com dados nulos

```python
df_apartamentos.isnull()
```

```python
df_apartamentos.isnull().sum()
```

```python
df_apartamentos = df_apartamentos.fillna(0)
```

```python
df_apartamentos.isnull().sum()
```

## Removendo registros

```python
df_apartamentos
```

```python
residencias_a_deletar = df_apartamentos.query('Condominio == 0 | Valor == 0').index
```

```python
df_apartamentos.drop(residencias_a_deletar, axis=0, inplace=True)
```

```python
df_apartamentos
```

```python
df_apartamentos.drop('Tipo', axis=1, inplace=True)
```

```python
df_apartamentos
```

## Filtros

### **1. Apartamentos que possuem `1 quarto` e `aluguel menor que 1200`**



```python
selecao1 = df_apartamentos['Quartos'] == 1
```

```python
df_apartamentos[selecao1]
```

```python
selecao2 = df_apartamentos['Valor'] < 1200
```

```python
df_apartamentos[selecao2]
```

```python
selecao_final = selecao1 & selecao2
```

```python
df_1 = df_apartamentos[selecao_final]
df_1
```

### **2. `Apartamentos` que possuem pelo menos `2 quartos`, `aluguel menor que 3000` e `area maior que 70`**

```python
selecoes = (df_apartamentos['Quartos'] >= 2) & (df_apartamentos['Valor'] < 3000) & (df_apartamentos['Area'] > 70)
```

```python
df_2 = df_apartamentos[selecoes]
df_2
```

# Manipulando os dados

## Criando colunas numéricas

```python
dados_numericos = pd.read_csv(url, sep=';')
dados_numericos
```

```python
dados_numericos['Valor_por_mes'] = dados_numericos['Valor'] + dados_numericos['Condominio']
dados_numericos
```

```python
dados_numericos['Valor_por_ano'] = dados_numericos['Valor_por_mes'] * 12 + dados_numericos['IPTU']
dados_numericos
```

## Criando colunas categóricas

```python
dados_numericos['Descricao'] = dados_numericos['Tipo'] + ' em ' + dados_numericos['Bairro'] + ' com ' + \
dados_numericos['Quartos'].astype(str) + ' quarto(s) ' + \
    ' e ' + dados['Vagas'].astype(str) + ' vaga(s) de garagem.'
```

```python
dados_numericos
```

```python
dados_numericos['Possui_suite'] = dados_numericos['Suites'].apply(lambda x: "Sim" if x > 0 else "Nao")
```

```python
dados_numericos
```

```python
dados_numericos.to_csv('dados_completos_dev.csv', index=False, sep=';')
```
