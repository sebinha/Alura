# Conhecendo a biblioteca Matplotlib

```python
import pandas as pd
```

```python
df_imigrantes_canada = pd.read_csv('imigrantes_canada.csv')
```

```python
df_imigrantes_canada
```

```python
df_imigrantes_canada.info()
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
df_imigrantes_canada.set_index('País', inplace=True)
```

```python
df_imigrantes_canada.head()
```

```python
df_imigrantes_canada.loc['Brasil', anos]
```

```python
Brasil = df_imigrantes_canada.loc['Brasil', anos]
```

```python
df_brasil = pd.DataFrame({'Anos': Brasil.index.tolist(), 'Imigrantes': Brasil.values.tolist()})
```

```python
df_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figsize=(10, 5)
plt.plot(df_brasil['Anos'], df_brasil['Imigrantes'])
plt.title('Imigrantes Brasileiros no Canadá de 1980 a 2013')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks([str(i) for i in range(1980, 2014, 5)])
plt.yticks([i for i in range(0, 3500, 500)])
plt.show()
```

# Criando figuras com Matplotlib

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df_brasil['Anos'], df_brasil['Imigrantes'], label='Brasil')
ax.set_title('Imigrantes Brasileiros no Canadá de\n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(10, 5))
axes[0].plot(df_brasil['Anos'], df_brasil['Imigrantes'], label='Brasil')
axes[0].set_title('Imigrantes Brasileiros no Canadá de\n 1980 a 2013')
axes[0].set_xlabel('Ano')
axes[0].set_ylabel('Número de Imigrantes')
axes[0].xaxis.set_major_locator(plt.MaxNLocator(5))

axes[1].boxplot(df_brasil['Imigrantes'])
axes[1].set_title('Imigrantes Brasileiros no Canadá de\n 1980 a 2013')
axes[1].set_ylabel('Número de Imigrantes')
axes[1].set_xticklabels(['Brasil']) 
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(10, 6))
fig.subplots_adjust(hspace=1, wspace=0.5)
fig.suptitle('Imigrantes dos quatro maiores países da América do Sul no Canadá de 1980 a 2013')
axes[0,0].plot(df_imigrantes_canada.loc['Brasil', anos], label='Brasil')
axes[0,1].plot(df_imigrantes_canada.loc['Argentina', anos], label='Argentina')
axes[1,0].plot(df_imigrantes_canada.loc['Colômbia', anos], label='Colômbia')
axes[1,1].plot(df_imigrantes_canada.loc['Peru', anos], label='Peru')

for ax in axes.flat:
    ax.set_title(ax.get_lines()[0].get_label())
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de Imigrantes')
    ax.xaxis.set_major_locator(plt.MaxNLocator(5))

ymin = 0
ymax = 7000

for ax in axes.flat:
    ax.set_ylim([ymin, ymax])

```

# Customizando com Matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(df_brasil['Anos'], df_brasil['Imigrantes'], lw=3)
ax.set_title('Imigração do Brasil para o Canadá\n1988 a 2013', fontsize=18, loc='left')
ax.plot(df_brasil['Anos'], df_brasil['Imigrantes'], lw=3, marker='o')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.grid(linestyle='--')
plt.show()
```

```python
df_america_do_sul = df_imigrantes_canada[df_imigrantes_canada['Região'] == 'América do Sul']
df_america_do_sul.sort_values(by='Total', ascending=True, inplace=True)
df_america_do_sul
```

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.barh(df_america_do_sul.index, df_america_do_sul['Total'], color=['skyblue', 'blue', 'green', 'red', 'yellow', 'orange', 'purple', 'brown', 'pink', 'gray'])
ax.set_title('Imigração da América do Sul para o Canadá\n1988 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('País', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
```

```python
fig, ax = plt.subplots(figsize=(10,6))

colors = ['green' if pais == 'Brasil' else 'silver' for pais in df_america_do_sul.index]

ax.barh(df_america_do_sul.index, df_america_do_sul['Total'], color=colors)

ax.set_title('América do Sul: Brasil está entre os quatro que \nmais imigraram para o Canadá de 1988 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

for i, v in enumerate(df_america_do_sul['Total']):
    ax.text(v, i, str(v), color='black', va='center', fontsize=12)
    
fig.savefig('imigrantes_america_do_sul.png', bbox_inches='tight', dpi=300)
```

```python
def gerar_grafico_paleta(pallete):
    fig, ax = plt.subplots(figsize=(10,6))
    ax = sns.barplot(data=df_top_10, x='Total', y=df_top_10.index, palette=pallete)
    ax.set(xlabel='Número de imigrantes', ylabel='País', title='Top 10 países que mais imigraram para o Canadá de 1988 a 2013')
    plt.show()
```

```python
import plotly.graph_objs as go
```

```python
import plotly.graph_objects as go

# Supondo que df_america_do_sul_final seja o DataFrame que você está usando
fig = go.Figure()

# Adicionando traços para Brasil e Argentina
fig.add_trace(
    go.Scatter(x=df_america_do_sul_final.index, y=df_america_do_sul_final['Brasil'], mode='lines', name='Brasil', line=dict(width=4))
)
fig.add_trace(
    go.Scatter(x=df_america_do_sul_final.index, y=df_america_do_sul_final['Argentina'], mode='lines', name='Argentina', line=dict(width=4))
)

# Atualizando o layout
fig.update_layout(
    title=dict(
        text='<b>Imigração do Brasil e Argentina para o Canadá no período de 1980 a 2013</b>',
        x=0.12,
        xanchor='left',
        font=dict(size=20)
    ),
    xaxis={'tickangle': -45},
    xaxis_title='Ano',
    yaxis_title='Número de Imigrantes',
    updatemenus=[dict(
        type='buttons',
        showactive=False,
        buttons=[dict(
            label='Play',
            method='animate',
            args=[None, {'frame': {'duration': 1000, 'redraw': True}, 'fromcurrent': True}]
        )]
    )],
    width=1000,
    height=500
)

# Criando os frames para a animação
frames = []
for i in range(len(df_america_do_sul_final)):
    frame_data = [
        go.Scatter(x=df_america_do_sul_final.index[:i+1], y=df_america_do_sul_final['Brasil'][:i+1], mode='lines', name='Brasil', line=dict(width=4)),
        go.Scatter(x=df_america_do_sul_final.index[:i+1], y=df_america_do_sul_final['Argentina'][:i+1], mode='lines', name='Argentina', line=dict(width=4))
    ]
    frame = go.Frame(data=frame_data)
    frames.append(frame)
fig.frames = frames

# Mostrando a figura
fig.show()
```
