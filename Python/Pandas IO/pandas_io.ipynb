# Trabalhando com banco de dados

```python
%pip install sqlalchemy
```

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:', isolation_level='AUTOCOMMIT')
```

```python
pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
```

```python
clientes = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv')
```

```python
clientes.to_sql('clientes', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados
```

```python
empregados.to_sql('empregados', engine, index=False)
```

```python
pd.read_sql_table('empregados', engine)
```

```python
pd.read_sql_table('empregados', engine, columns=["Idade", "Grau_escolaridade", "Rendimento_anual"])
```

```python
clientes
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
    conn.execute(text(query))
```

```python
pd.read_sql_table('clientes', engine)
```

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
    conn.execute(text(query))
```

```python
pd.read_sql_table('clientes', engine)
```
