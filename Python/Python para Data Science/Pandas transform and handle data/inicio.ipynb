```python
dados['preco']
```

```python
colunas_numericas_cifrao_float = ['preco', 'taxa_deposito', 'taxa_limpeza']
dados[colunas_numericas_cifrao_float] = dados[colunas_numericas_cifrao_float].applymap(lambda x: x.replace('$', '').replace(',', '').strip()).astype(np.float64)
dados.info()
```

# Dados textuais

```python
colunas_strings_hospedagem = ['descricao_local', 'descricao_vizinhanca']
dados[colunas_strings_hospedagem] = dados[colunas_strings_hospedagem].applymap(lambda x: str(x).lower())
dados
```

```python
dados.info()
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
```

```python
dados['descricao_local'].head()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"','',regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)
dados.head()
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```

```python
dados_vendas_clientes['Data de venda'] = pd.to_datetime(dados_vendas_clientes['Data de venda'], format='%d/%m/%Y')
dados_vendas_clientes.info()
```

```python
subset_vendas_dias = dados_vendas_clientes.groupby(dados_vendas_clientes['Cliente'])['Valor da compra'].sum()
subset_vendas_dias
```

```python
dados_locacao_imoveis['datas_combinadas_pagamento'] = pd.to_datetime(dados_locacao_imoveis['datas_combinadas_pagamento'], format='%d/%m/%Y')
dados_locacao_imoveis['datas_de_pagamento'] = pd.to_datetime(dados_locacao_imoveis['datas_de_pagamento'], format='%d/%m/%Y')
dados_locacao_imoveis.info()
```

```python
dados_locacao_imoveis['atraso'] = (dados_locacao_imoveis['datas_de_pagamento'] - dados_locacao_imoveis['datas_combinadas_pagamento']).dt.days
dados_locacao_imoveis
```

```python
media_atraso = dados_locacao_imoveis.groupby(['apartamento'])['atraso'].mean()
media_atraso
```
